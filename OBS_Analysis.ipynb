## ADVANCE QUARIES

### Q1. Retrieve the total number of books sold for each genre:

```python
query = """ SELECT B.GENRE, SUM(O.QUANTITY) FROM ORDERS AS O
JOIN BOOKS AS B ON O.BOOK_ID = B.BOOK_ID
GROUP BY GENRE; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Genre","Total_Books_Sold"])
```

### Q2. Find the average price of books in the "Fantasy" genre:

```python
query = """ SELECT GENRE, AVG(PRICE) FROM BOOKS
WHERE GENRE = "Fantasy"; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data, columns = ["Genre","Avg_Price"])
```

### Q3. List customers who have placed at least 2 orders:

```python
query = """ SELECT O.CUSTOMER_ID,C.NAME,COUNT(O.ORDER_ID)
FROM ORDERS AS O JOIN CUSTOMERS AS C ON O.CUSTOMER_ID = C.CUSTOMER_ID
GROUP BY O.CUSTOMER_ID,C.NAME HAVING COUNT(ORDER_ID)>=2; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Cust_ID","Name","ORDERS"])
```

### Q4. Find the most frequently ordered book:

```python
query = """ SELECT O.BOOK_ID, B.TITLE, COUNT(O.ORDER_ID) AS C
FROM ORDERS O JOIN BOOKS B ON O.BOOK_ID=B.BOOK_ID
GROUP BY O.BOOK_ID, B.TITLE ORDER BY C DESC LIMIT 1; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Book_ID","Title","Order_Count"])
```

### Q5. Show the top 3 most expensive books of 'Fantasy' Genre :

```python
query = """ SELECT *FROM BOOKS WHERE GENRE = "Fantasy"
ORDER BY PRICE DESC LIMIT 3; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Book_ID","Title","Auther","Genre","Published_Year","Price","Stock"])
```

### Q6. Retrieve the total quantity of books sold by each author:

```python
query = """ SELECT B.Auther, SUM(O.QUANTITY) FROM BOOKS B
JOIN ORDERS O ON B.BOOK_ID=O.BOOK_ID
GROUP BY B.Auther ORDER BY B.Auther; """ 
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Author","Quantity"])
```

### Q7. List the cities where customers who spent over $30 are located:

```python
query = """ SELECT C.CITY, SUM(O.TOTAL_AMOUNT) FROM CUSTOMERS C 
JOIN ORDERS O ON C.CUSTOMER_ID=O.CUSTOMER_ID
WHERE O.TOTAL_AMOUNT>30 GROUP BY C.CITY
ORDER BY C.CITY; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["City","Total_Amount"])
```

### Q8. Find the customer who spent the most on orders:

```python
query = """ SELECT C.CUSTOMER_ID,C.NAME,C.EMAIL,SUM(O.TOTAL_AMOUNT) AS T 
FROM CUSTOMERS C JOIN ORDERS O GROUP BY C.CUSTOMER_ID, C.NAME
ORDER BY T DESC LIMIT 1; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Cust_ID","Name","Email","Total_Amount"])
```

### Q9. Calculate the stock remaining after fulfilling all orders:

```python
query = """ SELECT B.BOOK_ID,B.TITLE,B.STOCK,COALESCE(SUM(O.QUANTITY),0),
B.STOCK-COALESCE(SUM(O.QUANTITY),0) AS REMAINING_STOCK
FROM BOOKS B JOIN ORDERS O ON B.BOOK_ID=O.BOOK_ID
GROUP BY B.BOOK_ID ORDER BY B.BOOK_ID; """
cur.execute(query)
data = cur.fetchall()
pd.DataFrame(data,columns=["Book_ID","Title","Stock","Order_Quantity","Remaining_Stock"])
```
